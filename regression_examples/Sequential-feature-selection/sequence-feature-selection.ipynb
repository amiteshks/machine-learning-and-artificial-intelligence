## ðŸ“Š Project: Linear Regression Model using Sequential Feature Selection, 3rd Degree Polynomial Features and Scikit Pipeline 

This Jupyter Notebook demonstrates a step-by-step workflow for building a Linear Regression model using Scikit-learn pipelines. Below is the flow of the program:

1. **Load Data**  
   - Import the dataset and perform initial data exploration.

2. **Split Data into Train and Test Sets**  
   - Divide the data into training and testing sets to evaluate the model's performance.

  

3. **Create a New Dataset with 3rd Degree Polynomial Features**  
   - Use polynomial feature transformation to capture non-linear relationships in the data.

4. **Feature Selection**  
   - Select the **best 3 features** based on minimizing the Mean Squared Error (MSE).

5. **Build and Train a Linear Regression Model**  
   - Use the selected features to build a **Linear Regression model** within a Scikit-learn pipeline for efficient preprocessing and modeling.


```python
### Import sklearn libraries
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline
from sklearn import set_config
set_config(display="diagram")

### Import matplotlib, pandas and numpy libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Load Data

The [Medical Cost Personal dataset](https://www.kaggle.com/mirichoi0218/insurance) used in this assignment is from Kaggle and contains information on insurance clients and their medical costs billed by the insurance company.  We will build regression models to predict the logarithm of the `charges` column. The dataset is loaded and displayed below, and the original target is plotted next to its resulting logarithm.   

```python
#Load Insurance Data
insurance = pd.read_csv('data/insurance.csv')
insurance.head()
```

### Split Data into Train and Test Sets

```python
# Get X_train, X_test, y_train, y_test of the data
insurance[['log_charges']] = np.log1p(insurance[['charges']])
X = insurance.drop(['charges'], axis=1)
y = X['log_charges']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Create a New Dataset with 3rd Degree Polynomial Features

```python

#Create a Polynomial Feature of degree 3
poly_features = PolynomialFeatures(degree=3, include_bias=False)

#Create Train dataframe based on the Polynomial features
labels = ['age', 'bmi', 'children']

X_train_poly = poly_features.fit_transform(X_train[labels])
train_df = pd.DataFrame(X_train_poly, columns = poly_features.get_feature_names_out(labels))

#Create Test dataframe based on the Polynomial features
X_test_poly = poly_features.fit_transform(X_test[labels])
test_df = pd.DataFrame(X_test_poly, columns = poly_features.get_feature_names_out(labels))

train_df.head()
```

### Feature Selection

Select the best 3 features based on minimizing the Mean Squared Error (MSE).

```python


selector = ''
best_features = ''
best_features_df = ''

selector = SequentialFeatureSelector(LinearRegression(), n_features_to_select=3)
best_features = selector.fit_transform(train_df, y_train)
best_features_df = pd.DataFrame(best_features, columns=selector.get_feature_names_out())
                                

# ANSWER CHECK
best_features_df.head()
```

### `Pipeline` for regression model

Build and Train a Linear Regression Model


```python


pipe = ''
train_preds = ''
test_preds = ''
train_mse = ''
test_mse = ''

linreg = LinearRegression()

pipe = Pipeline([('column_selector', selector),
                 ('regression', linreg)])

# Fit the training data
pipe.fit(train_df, y_train)

#Predict training and test data
train_preds = pipe.predict(train_df)
test_preds = pipe.predict(test_df)

#Calculate mean squared error
train_mse = mean_squared_error(train_preds, y_train)
test_mse = mean_squared_error(test_preds, y_test)
print ('train_mse', train_mse)


# Answer check
print(f'Train MSE: {train_mse: .2f}')
print(f'Test MSE: {test_mse: .2f}')
pipe
```

### Backward Selection Pipeline for regression model

Build and Train a Linear Regression Mode using Backward Selection Pipelinel

```python
backward_selector = SequentialFeatureSelector(LinearRegression(), 
                                              n_features_to_select=3,
                                              direction = 'backward')
```

```python

backward_pipe = Pipeline([('selector', backward_selector), 
                         ('regression', LinearRegression())])

backward_pipe.fit(train_df, y_train)
train_preds_backward = backward_pipe.predict(train_df)
backward_train_mse = mean_squared_error(train_preds_backward, y_train)
backward_test_mse = mean_squared_error(backward_pipe.predict(test_df), y_test)


# Answer check
print(f'Train MSE: {backward_train_mse: .2f}')
print(f'Test MSE: {backward_test_mse: .2f}')
backward_pipe
```
