# Semantic Search on a Small Knowledge Base

```python
import numpy as np
documents = [
    "You can reset your password by going to the login page.",
    "Update your email address under account settings.",
    "To contact support, click on the Help Center.",
    "You can request a refund within 30 days of purchase.",
    "To change your subscription, visit the billing section."
]

# Generate embeddings for the documents
doc_response = client.embeddings.create(
    model="text-embedding-ada-002",
    input=documents
)
doc_embeddings = np.array([d.embedding for d in doc_response.data])
```

```python
# User query
query = "How do I get my money back?"

# Generate embedding for the query
query_response = client.embeddings.create(
    model="text-embedding-ada-002",
    input=[query]
)
query_embedding = np.array(query_response.data[0].embedding).reshape(1, -1)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# Compute cosine similarity
similarities = cosine_similarity(query_embedding, doc_embeddings)[0]

# Rank and show top results
top_idx = np.argsort(similarities)[::-1]
print("Top 3 matching documents:")
for idx in top_idx[:3]:
    print(f"- Score: {similarities[idx]:.4f} | {documents[idx]}")
```
